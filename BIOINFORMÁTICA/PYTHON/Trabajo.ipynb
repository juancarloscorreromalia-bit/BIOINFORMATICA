#### 6.- Realiza una función que realice una digestión con las siguientes enzimas de restricción en nuestra secuencia. La función tiene que devolver la siguiente información: (2 pto)
1. Enzimas que tienen sitios de corte en la secuencia y posición (o posiciones de corte)

```python
enzR = {
    'HindIII' : 'A*AGCTT',
    'BamHI' : 'G*GATCC',
    'AluI' : 'AG*CT',
    'Sau3AI' : '*GATC',
    'EcoRI' : 'G*AATTC'
}
```

```python
enzR_sin={}
for nombre_enz in enzR:
    sec = enzR[nombre_enz]
    enzR_sin[nombre_enz] = sec.replace("*","")
print(enzR_sin)
```

```python
def Digestion_enz(secuencia , enzR_sin): 
    resultado = {}
    for nombre, secc in enzR_sin.items():
        posicion = [match.start() for match in re.finditer(secc,secuencia)]
        if posicion: 
            resultado[nombre] = posicion
    return resultado
```

```python
Resultados_digestion = Digestion_enz(secuencia, enzR_sin)
print(Resultados_digestion)
```

```python
def No_Digestion_enz(secuencia , enzR_sin): 
    sin_corte = []
    for nombre_2, secc_2 in enzR_sin.items():
        posicion_2 = [match.start() for match in re.finditer(secc_2,secuencia)]
        if len(posicion_2) == 0:
            sin_corte.append(nombre_2)
    return sin_corte
```

```python
Enz_no_funcionantes = No_Digestion_enz(secuencia, enzR_sin)
print(Enz_no_funcionantes)
```

```python
def Fragmentos_enz(secuencia, posicion):
    if not posicion:
        return [len(secuencia)]
    cortes = sorted(posicion)
    puntos_de_corte = [0] + cortes + [len(secuencia)]
    tamaños = []
    for e in range(len(puntos_de_corte) - 1):
        inicio = puntos_de_corte[e]
        fin = puntos_de_corte[e + 1]
        tamaño = fin - inicio 
        tamaños.append(tamaño) 
    return tamaños
```

```python
for enz, posc in Resultados_digestion.items():
    frags = Fragmentos_enz(secuencia, posc)
    print(f"La enzima {enz} posee {len(frags)} fragmentos con los siguientes tamaños: {frags}")
```
