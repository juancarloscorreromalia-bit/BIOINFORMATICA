#### 7.- Con la secuencia y la lista siguiente, escribe un algoritmo que cuente el número de veces que aparece un elemento de la lista en la secuencia. Pista: hacer un diccionario para almacenar los resultados puede ser buena idea

```python
dna = "AATGATGAACGAC" 
dinucleotides = ['AA','AT','AG','AC', 
                 'TA','TT','TG','TC', 
                 'GA','GT','GG','GC', 
                 'CA','CT','CG','CT'] 
```

```python

mi_dicc={} 
for elemento in dinucleotides: 
    contaje = dna.count(elemento)
    mi_dicc[elemento]= contaje
print(mi_dicc)
# Hacemos un bucle for, donde por cada elemento de la lista de dinucleótidos, en dna haga un contaje y 
# luego creamos un diccionario que atribuya a cada dinuclóotido los valores en el contaje de dna. 
```

##### ¿Cuántos counts tiene TC?

```python
mi_dicc["TC"] # Por medio del Key, sacamos el valor que es el numero de veces que esta en la secuencia dna 
```

##### ¿Puedes imprimir los dinuleótidos que tengan counts igual a 2?

```python
for elemento in mi_dicc:
    if mi_dicc[elemento] == 2:
        print(elemento)
# Hacemos un bucle for, donde por cada dinucleótido(key) mire el número de counts(value) 
```
