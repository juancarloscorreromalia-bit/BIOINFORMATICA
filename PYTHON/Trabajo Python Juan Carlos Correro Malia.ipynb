```python
def Digestion_enz(secuencia , enzR_sin): 
    resultado = {}
    for nombre, secc in enzR_sin.items():
        posicion = [match.start() for match in re.finditer(secc,secuencia)]
        if posicion: 
            resultado[nombre] = posicion
    return resultado
```

```python
def Fragmentos_enz(secuencia, posicion):
    if not posicion:
        return [len(secuencia)]
    cortes = sorted(posicion)
    puntos_de_corte = [0] + cortes + [len(secuencia)]
    tamaños = []
    for e in range(len(puntos_de_corte) - 1):
        inicio = puntos_de_corte[e]
        fin = puntos_de_corte[e + 1]
        tamaño = fin - inicio 
        tamaños.append(tamaño) 
    return tamaños
```
