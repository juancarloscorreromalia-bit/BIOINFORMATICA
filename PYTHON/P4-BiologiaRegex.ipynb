##### 4.- Escribe una función que cuente el porcentaje de A de una secuencia

```python
def porcentaje_a(genérico):
    porc_a = []
    for a in genérico: 
        if a == "A": 
            porc_a.append(a)
    longuitud_a = len(porc_a)
    longuitud_total = len(genérico)
    porcentaje =  (longuitud_a/longuitud_total) * 100 
    return porcentaje
# Creamos una funcion y dentro de ella creamos una lista vacia donde se almacenarán las "A".
# Luego creamos un bucle que busque estas en la secuencia y que las meta en la lista creada anteriormente.
# Posteriormente medimos la longuitud de la lista A y de la secuencia total, y calculamos el porcentaje.
# Hacemos que el bucle nos devuelva el porcentaje.
```

##### ¿Puedes hacer que devuelva dos decimales? ¿Y que lidie con mayúsculas y minúsculas?

```python
def porcentaje_a(genérico):
    porc_a = []
    for a in genérico: 
        if a == "a" or a == "A": 
            porc_a.append(a)
    longuitud_a = len(porc_a)
    longuitud_total = len(genérico)
    porcentaje =  (longuitud_a/longuitud_total) * 100 
    return round(porcentaje, 2)
# Igual que la anterior pero ponemos que busque tanto "a" como "A", y le decimos que
# devuelva el resultado redondeado por medio de la funcion round(,2) a dos decimales.
```
